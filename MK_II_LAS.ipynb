```python
all_data_LAS=pd.DataFrame()

for i in range(0,len(file_list)):
    print file_list[i]

    RecordNo=[]
    dt=[]
    all_date=[]
    StatusFlags=[]
    Udemod=[]
    UdemodSig=[]
    Cn2=[]
    Cn2Sig=[]
    Cn2Min=[]
    Cn2Max=[]
    Srt=[]
    Hfree=[]
    DeviceTemp=[]
    AirTemp=[] 
    AirPressure=[]
    WindSpeed=[]

    dt_LAS=pd.DataFrame()

#i=300
    with open(file_list[i],'rt') as f:
        for count,line in enumerate(f):
            if count>1: # To skip first 2 rows of data
             dt_field=line.split('\t')
             date=str(dt_field[0])
             yr=date[0:4] 
             mon=date[5:7]
             day=date[8:10] 
             hr=date[11:13]
             min=date[14:16]
             sec=date[17:20]
             dt=datetime.datetime(int(yr),int(mon),int(day),int(hr),int(min),int(sec))

             all_date.append(dt)   
             RecordNo.append(int(dt_field[1]))
             StatusFlags.append(int(dt_field[2]))
             Udemod.append(float(dt_field[3])) 
             UdemodSig.append(float(dt_field[4]))
             Cn2.append(float(dt_field[5]))
             Cn2Sig.append(float(dt_field[6]))
             Cn2Min.append(float(dt_field[7]))
             Cn2Max.append(float(dt_field[8]))
             Srt.append(float(dt_field[9]))
             Hfree.append(float(dt_field[10]))
             DeviceTemp.append(float(dt_field[11]))
             AirTemp.append(float(dt_field[12])) 
             AirPressure.append(float(dt_field[13]))
             WindSpeed.append(float(dt_field[14]))

    dt_LAS['RecordNo']=RecordNo  
    dt_LAS['StatusFlags']=StatusFlags 
    dt_LAS['Udemod']=Udemod 
    dt_LAS['UdemodSig']=UdemodSig
    dt_LAS['Cn2']=Cn2 
    dt_LAS['Cn2Sig']=Cn2Sig
    dt_LAS['Cn2Min']=Cn2Min
    dt_LAS['Cn2Max']=Cn2Max
    dt_LAS['Srt']=Srt
    dt_LAS['Hfree']=Hfree
    dt_LAS['DeviceTemp']=DeviceTemp
    dt_LAS['AirTemp']=AirTemp
    dt_LAS['AirPressure']=AirPressure
    dt_LAS['WindSpeed']=WindSpeed

    dt_LAS.index=all_date

    #print dt_LAS
    all_data_LAS=all_data_LAS.append(dt_LAS)
```

```python
all_data_LAS_f=all_data_LAS[all_data_LAS['Hfree']<600]

all_data_LAS_f1=all_data_LAS_f[all_data_LAS['Udemod']<-400]

LAS_30min=all_data_LAS_f1.resample('30min',label='right', closed='left').mean()

LAS_D=LAS_30min.resample('D',label='right', closed='left').mean()
```
